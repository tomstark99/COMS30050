```python
import seaborn as sns
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
all_targets = pd.read_csv('../data/0&1/y_targets.csv')
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

X_ds = all_targets.copy()
ds = X_ds.corr()
sns.heatmap(ds, annot=True)

ax.set_title('depression indicators', fontsize=20)
```
