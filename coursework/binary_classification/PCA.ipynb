```python
from sklearn import datasets
from matplotlib import pyplot as plt

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# PCA 
from sklearn.decomposition import KernelPCA
```

# Dep score

```python
total_dep_score = pd.read_csv('../data/0&1/total_dep_score.csv')
sampled_dep_score = pd.read_csv('../data/0&1/sampled_dep_score.csv')

# X 
X_t = total_dep_score.copy()
del X_t['dep_score']

X_s = sampled_dep_score.copy()
del X_s['dep_score']

# y 
y_t = total_dep_score['dep_score']
y_s = sampled_dep_score['dep_score']

# total 
linear_t  = KernelPCA(n_components=2, random_state=1, kernel='linear')
poly_t    = KernelPCA(n_components=2, random_state=1, kernel='poly')
rbf_t     = KernelPCA(n_components=2, random_state=1, kernel='rbf')
sigmoid_t = KernelPCA(n_components=2, random_state=1, kernel='sigmoid')
cosine_t  = KernelPCA(n_components=2, random_state=1, kernel='cosine')

linear_X_t  = linear_t.fit_transform(X_t)
poly_X_t    = poly_t.fit_transform(X_t)
rbf_X_t     = rbf_t.fit_transform(X_t)
sigmoid_X_t = sigmoid_t.fit_transform(X_t)
consine_X_t = cosine_t.fit_transform(X_t)

# sampled 
linear_s  = KernelPCA(n_components=2, random_state=1, kernel='linear')
poly_s    = KernelPCA(n_components=2, random_state=1, kernel='poly')
rbf_s     = KernelPCA(n_components=2, random_state=1, kernel='rbf')
sigmoid_s = KernelPCA(n_components=2, random_state=1, kernel='sigmoid')
cosine_s  = KernelPCA(n_components=2, random_state=1, kernel='cosine')

linear_X_s  = linear_s.fit_transform(X_s)
poly_X_s    = poly_s.fit_transform(X_s)
rbf_X_s     = rbf_s.fit_transform(X_s)
sigmoid_X_s = sigmoid_s.fit_transform(X_s)
consine_X_s = cosine_s.fit_transform(X_s)

fig, ((ax1, ax2, ax3, ax4, ax5), (ax6, ax7, ax8, ax9, ax10)) = plt.subplots(2, 5, figsize=(25,10))

for l in np.unique(y_t):
    ax1.scatter(linear_X_t[y_t==l, 0],  linear_X_t[y_t==l, 1],  label=l, alpha=0.5)
    ax2.scatter(poly_X_t[y_t==l, 0],    poly_X_t[y_t==l, 1],    label=l, alpha=0.5)
    ax3.scatter(rbf_X_t[y_t==l, 0],     rbf_X_t[y_t==l, 1],     label=l, alpha=0.5)
    ax4.scatter(sigmoid_X_t[y_t==l, 0], sigmoid_X_t[y_t==l, 1], label=l, alpha=0.5)
    ax5.scatter(consine_X_t[y_t==l, 0], consine_X_t[y_t==l, 1], label=l, alpha=0.5)
    
    ax6.scatter(linear_X_s[y_s==l, 0],  linear_X_s[y_s==l, 1],  label=l, alpha=0.5)
    ax7.scatter(poly_X_s[y_s==l, 0],    poly_X_s[y_s==l, 1],    label=l, alpha=0.5)
    ax8.scatter(rbf_X_s[y_s==l, 0],     rbf_X_s[y_s==l, 1],     label=l, alpha=0.5)
    ax9.scatter(sigmoid_X_s[y_s==l, 0], sigmoid_X_s[y_s==l, 1], label=l, alpha=0.5)
    ax10.scatter(consine_X_s[y_s==l, 0], consine_X_s[y_s==l, 1], label=l, alpha=0.5)

ax5.legend()
ax10.legend()

ax1.set_title('Linear Kernel (Total)')
ax2.set_title('Poly Kernel (Total)')
ax3.set_title('RBF Kernel (Total)')
ax4.set_title('Sigmoid Kernel (Total)')
ax5.set_title('Cosine Kernel (Total)')

ax6.set_title('Linear Kernel (Sampled)')
ax7.set_title('Poly Kernel (Sampled)')
ax8.set_title('RBF Kernel (Sampled)')
ax9.set_title('Sigmoid Kernel (Sampled)')
ax10.set_title('Cosine Kernel (Sampled)')
```

```python
total_panic_score = pd.read_csv('../data/0&1/total_panic_score.csv')
sampled_panic_score = pd.read_csv('../data/0&1/sampled_panic_score.csv')

# X 
X_t = total_panic_score.copy()
del X_t['panic_score']

X_s = sampled_panic_score.copy()
del X_s['panic_score']

# y 
y_t = total_panic_score['panic_score']
y_s = sampled_panic_score['panic_score']

# total 
linear_t  = KernelPCA(n_components=2, random_state=1, kernel='linear')
poly_t    = KernelPCA(n_components=2, random_state=1, kernel='poly')
rbf_t     = KernelPCA(n_components=2, random_state=1, kernel='rbf')
sigmoid_t = KernelPCA(n_components=2, random_state=1, kernel='sigmoid')
cosine_t  = KernelPCA(n_components=2, random_state=1, kernel='cosine')

linear_X_t  = linear_t.fit_transform(X_t)
poly_X_t    = poly_t.fit_transform(X_t)
rbf_X_t     = rbf_t.fit_transform(X_t)
sigmoid_X_t = sigmoid_t.fit_transform(X_t)
consine_X_t = cosine_t.fit_transform(X_t)

# sampled 
linear_s  = KernelPCA(n_components=2, random_state=1, kernel='linear')
poly_s    = KernelPCA(n_components=2, random_state=1, kernel='poly')
rbf_s     = KernelPCA(n_components=2, random_state=1, kernel='rbf')
sigmoid_s = KernelPCA(n_components=2, random_state=1, kernel='sigmoid')
cosine_s  = KernelPCA(n_components=2, random_state=1, kernel='cosine')

linear_X_s  = linear_s.fit_transform(X_s)
poly_X_s    = poly_s.fit_transform(X_s)
rbf_X_s     = rbf_s.fit_transform(X_s)
sigmoid_X_s = sigmoid_s.fit_transform(X_s)
consine_X_s = cosine_s.fit_transform(X_s)

fig, ((ax1, ax2, ax3, ax4, ax5), (ax6, ax7, ax8, ax9, ax10)) = plt.subplots(2, 5, figsize=(25,10))

for l in np.unique(y_t):
    ax1.scatter(linear_X_t[y_t==l, 0],  linear_X_t[y_t==l, 1],  label=l, alpha=0.5)
    ax2.scatter(poly_X_t[y_t==l, 0],    poly_X_t[y_t==l, 1],    label=l, alpha=0.5)
    ax3.scatter(rbf_X_t[y_t==l, 0],     rbf_X_t[y_t==l, 1],     label=l, alpha=0.5)
    ax4.scatter(sigmoid_X_t[y_t==l, 0], sigmoid_X_t[y_t==l, 1], label=l, alpha=0.5)
    ax5.scatter(consine_X_t[y_t==l, 0], consine_X_t[y_t==l, 1], label=l, alpha=0.5)
    
    ax6.scatter(linear_X_s[y_s==l, 0],  linear_X_s[y_s==l, 1],  label=l, alpha=0.5)
    ax7.scatter(poly_X_s[y_s==l, 0],    poly_X_s[y_s==l, 1],    label=l, alpha=0.5)
    ax8.scatter(rbf_X_s[y_s==l, 0],     rbf_X_s[y_s==l, 1],     label=l, alpha=0.5)
    ax9.scatter(sigmoid_X_s[y_s==l, 0], sigmoid_X_s[y_s==l, 1], label=l, alpha=0.5)
    ax10.scatter(consine_X_s[y_s==l, 0], consine_X_s[y_s==l, 1], label=l, alpha=0.5)

ax5.legend()
ax10.legend()

ax1.set_title('Linear Kernel (Total)')
ax2.set_title('Poly Kernel (Total)')
ax3.set_title('RBF Kernel (Total)')
ax4.set_title('Sigmoid Kernel (Total)')
ax5.set_title('Cosine Kernel (Total)')

ax6.set_title('Linear Kernel (Sampled)')
ax7.set_title('Poly Kernel (Sampled)')
ax8.set_title('RBF Kernel (Sampled)')
ax9.set_title('Sigmoid Kernel (Sampled)')
ax10.set_title('Cosine Kernel (Sampled)')
```
