```python
import seaborn as sns
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
total_dep_score = pd.read_csv('../data/0&1/sampled_dep_score.csv')
total_dep_thoughts = pd.read_csv('../data/0&1/sampled_dep_thoughts.csv')
total_has_dep_diag = pd.read_csv('../data/0&1/sampled_has_dep_diag.csv')
total_prim_diag = pd.read_csv('../data/0&1/sampled_prim_diag.csv')
total_secd_diag = pd.read_csv('../data/0&1/sampled_secd_diag.csv')
```

# Total

## Dep score

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1, 5, figsize=(50, 7))

X_ds = total_dep_score.copy()
ds = X_ds.corr()
sns.heatmap(ds, annot=True, ax=ax1)

X_dt = total_dep_thoughts.copy()
dt = X_dt.corr()
sns.heatmap(dt, annot=True, ax=ax2)

X_hdp = total_has_dep_diag.copy()
hdp = X_hdp.corr()
sns.heatmap(hdp, annot=True, ax=ax3)

X_pd = total_prim_diag.copy()
pd = X_pd.corr()
sns.heatmap(pd, annot=True, ax=ax4)

X_sd = total_secd_diag.copy()
sd = X_sd.corr()
sns.heatmap(sd, annot=True, ax=ax5)

ax1.set_title('dep_score', fontsize=20)
ax2.set_title('dep_thoughts', fontsize=20)
ax3.set_title('has_dep_diag', fontsize=20)
ax4.set_title('prim_diag', fontsize=20)
ax5.set_title('secd_diag', fontsize=20)
```
