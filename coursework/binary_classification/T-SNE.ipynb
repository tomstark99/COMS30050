```python
from sklearn import datasets
from matplotlib import pyplot as plt

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# T-SNE 
from sklearn.manifold import TSNE
```

# Divided into two sections: total and sampled

#### Total: The data is converted to binary classification across all Y targets 

#### Sampled: Similar to 'total' but there is a minor difference. The data is evenly distributed. For example, if 'dep_score depression=0' counts 100 and 'dep_score depression=1' counts 60 then the sampled data looks like: 'dep_score depression=0' counts 60 and 'dep_score depression=1' counts 60. The number of 'dep_score depression=0' changed from 100 to 60 so that the number of data for each category(depression=0 or 1) is even. 


- Y targets: dep_score, dep_thoughts, has_dep_diag, prim_diag, secd_diag
- 0: not depressed
- 1: depressed

# Total dep score

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1, 5, figsize=(35, 7))

# dep score
total_dep_score = pd.read_csv('../data/0&1/total_dep_score.csv')

X = total_dep_score.copy()
del X['dep_score']

target = total_dep_score['dep_score']

tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)
target_ids = range(2)

plt.figure(figsize=(6, 6))
colors = 'r', 'g'
for i, c, label in zip(target_ids, colors, ['not depressed', 'depressed']):
    ax1.scatter(X_tsne[target == i, 0], X_tsne[target == i, 1], c=c, label=label, alpha=0.5)

# dep thoughts 
total_dep_thoughts = pd.read_csv('../data/0&1/total_dep_thoughts.csv')

X = total_dep_thoughts.copy()
del X['dep_thoughts']

target = total_dep_thoughts['dep_thoughts']

tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)
target_ids = range(2)

plt.figure(figsize=(6, 6))
colors = 'r', 'g'
for i, c, label in zip(target_ids, colors, ['not depressed', 'depressed']):
    ax2.scatter(X_tsne[target == i, 0], X_tsne[target == i, 1], c=c, label=label, alpha=0.5)

# has_dep_diag
total_has_dep_diag = pd.read_csv('../data/0&1/total_has_dep_diag.csv')

X = total_has_dep_diag.copy()
del X['has_dep_diag']

target = total_has_dep_diag['has_dep_diag']

tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)
target_ids = range(2)

plt.figure(figsize=(6, 6))
colors = 'r', 'g'
for i, c, label in zip(target_ids, colors, ['not depressed', 'depressed']):
    ax3.scatter(X_tsne[target == i, 0], X_tsne[target == i, 1], c=c, label=label, alpha=0.5)

# prim_diag 
total_prim_diag = pd.read_csv('../data/0&1/total_prim_diag.csv')

X = total_prim_diag.copy()
del X['prim_diag']

target = total_prim_diag['prim_diag']

tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)
target_ids = range(2)

plt.figure(figsize=(6, 6))
colors = 'r', 'g'
for i, c, label in zip(target_ids, colors, ['not depressed', 'depressed']):
    ax4.scatter(X_tsne[target == i, 0], X_tsne[target == i, 1], c=c, label=label, alpha=0.5)

# secd_diag
total_secd_diag = pd.read_csv('../data/0&1/total_secd_diag.csv')

X = total_secd_diag.copy()
del X['secd_diag']

target = total_secd_diag['secd_diag']

tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)
target_ids = range(2)

plt.figure(figsize=(6, 6))
colors = 'r', 'g'
for i, c, label in zip(target_ids, colors, ['not depressed', 'depressed']):
    ax5.scatter(X_tsne[target == i, 0], X_tsne[target == i, 1], c=c, label=label, alpha=0.5)
ax5.legend(loc='upper center', bbox_to_anchor=(-1.9, -0.1),
          fancybox=True, shadow=True, ncol=2)


ax1.set_title('dep_score')
ax2.set_title('dep_thoughts')
ax3.set_title('has_dep_diag')
ax4.set_title('prim_diag')
ax5.set_title('secd_diag')

plt.show()
```
