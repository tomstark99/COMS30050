```python
from sklearn import datasets
from matplotlib import pyplot as plt

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# T-SNE 
from sklearn.manifold import TSNE
```

# Divided into two sections: total and sampled

#### Total: The data is converted to binary classification across all Y targets 

#### Sampled: Similar to 'total' but there is a minor difference. The data is evenly distributed. For example, if 'dep_score depression=0' counts 100 and 'dep_score depression=1' counts 60 then the sampled data looks like: 'dep_score depression=0' counts 60 and 'dep_score depression=1' counts 60. The number of 'dep_score depression=0' changed from 100 to 60 so that the number of data for each category(depression=0 or 1) is even. 


- Y targets: dep_score, dep_thoughts, has_dep_diag, prim_diag, secd_diag
- 0: not depressed
- 1: depressed

# Total dep score

```python
total_dep_score = pd.read_csv('../data/0&1/total_dep_score.csv')

X = total_dep_score.copy()
del X['dep_score']

target = total_dep_score['dep_score']

tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)
target_ids = range(2)

plt.figure(figsize=(6, 6))
colors = 'r', 'g'
for i, c, label in zip(target_ids, colors, ['not depressed', 'depressed']):
    plt.scatter(X_tsne[target == i, 0], X_tsne[target == i, 1], c=c, label=label)
plt.legend()
plt.show()
```
