```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
filename = 'sum_targets_with_panic.csv'
data = pd.read_csv(filename)
```

```python
data
```

```python
data_numbers = {}

data_numbers['total'] = dict(sorted(data['total'].value_counts().items()))
```

```python
data_numbers['total']
```

```python
x = np.arange(len(data_numbers['total']))
values = [data_numbers['total'][i] for i in data_numbers['total']]

plt.bar(x, values)
plt.xticks(x, data_numbers['total'].keys())
plt.title("The number of 'total' values")
plt.show()
```
