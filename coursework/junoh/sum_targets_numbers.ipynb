```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
filename = '../data/0&1/sum_targets_with_panic.csv'
data = pd.read_csv(filename)
```

```python
data
```

```python
data_numbers = {}

data_numbers['total'] = dict(sorted(data['total'].value_counts().items()))

zero = { 0.0: 0 }

zero.update(data_numbers['total'])
```

```python
data_numbers = zero
```

```python
x = np.arange(len(data_numbers))
values = [data_numbers[i] for i in data_numbers]

plt.bar(x, values)
plt.xticks(x, data_numbers.keys())
plt.title("The number of 'total' values")
plt.show()
```
