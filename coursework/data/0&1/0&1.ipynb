```python
minimise_target = {
    'dep_score': 0.0, # num=704
    'dep_thoughts': 1.0, # num=62
    'has_dep_diag': 0, # num=242
    'prim_diag': 0.0, # num=613
    'secd_diag': 0.0, # num=322
    'panic_score':0.0, # num=60
    
}



final_outcome = {
    
}
```

```python
for x in targets:
    num = 0
    
    # Copy 
    copied_dict_all = dict_all[x].copy()
    tem = dict_all[x].copy()
    
    tem = tem.loc[tem[x] == minimise_target[x]]
    
    if x == 'dep_score':
        num = 704
    elif x == 'dep_thoughts':
        num = 62
    elif x == 'has_dep_diag':
        num = 242
    elif x == 'prim_diag':
        num = 613
    elif x == 'secd_diag':
        num = 322
    elif x == "panic_score":
        num = 60
    else:
        pass
    
    final_outcome[x] = tem.sample(n=num).append(copied_dict_all.loc[copied_dict_all[x] != minimise_target[x]])
    final_outcome[x] = final_outcome[x].sample(frac=1)
```

```python
sampled_dep_score = final_outcome['dep_score']
sampled_dep_thoughts = final_outcome['dep_thoughts']
sampled_has_dep_diag = final_outcome['has_dep_diag']
sampled_prim_diag = final_outcome['prim_diag']
sampled_secd_diag = final_outcome['secd_diag']
sampled_panic_score = final_outcome['panic_score']
```
