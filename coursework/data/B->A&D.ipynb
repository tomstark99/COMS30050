```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Change the filename 
filename = 'processed_data.csv'
data = pd.read_csv(filename)
```

```python
data
```

```python
dict_all = {
    
}
```

```python
B_comp = ['comp_house', 'comp_int_bed_16', 'comp_noint_bed_16']

B_text = ['tran_week', 'tran_wend']

A_comp = ['comp_week', 'comp_wend']

A_text = ['text_week', 'text_wend']

```

```python
# B_comp
for x in B_comp:
    features = data[[x]].copy()
    features.loc[features[x] == 'No', x] = 0
    features.loc[features[x] == 'Yes', x] = 1
    
    if x == 'comp_house':
        pd_comp = features
    else:
        pd_comp[x] = features
```

```python
# B_text
for x in B_text:
    features = data[[x]].copy()
    features.loc[features[x] == 'Less than 1 hour', x] = 0
    features.loc[features[x] == '1 or more hours', x] = 1
    
    if x == 'tran_week':
        pd_text = features
    else:
        pd_text[x] = features
```

```python
pd_comp
```

```python
pd_text
```

```python
# comp
for x in A_comp:
    features = data[[x]].copy()
    features.loc[features[x] == 'Not at all', x] = 0
    features.loc[features[x] == 'Less than 1 hour', x] = 1
    features.loc[features[x] == '1-2 hours', x] = 2
    features.loc[features[x] == '3 or more hours', x] = 3
    
    if x == 'comp_week':
        pd_comp_targets = features
    else:
        pd_comp_targets[x] = features
```

```python
# text_week
for x in A_text:
    features = data[[x]].copy()
    features.loc[features[x] == 'Not at all', x] = 0
    features.loc[features[x] == 'Less than 1 hour', x] = 1
    features.loc[features[x] == '1-2 hours', x] = 2
    features.loc[features[x] == '3 or more hours', x] = 3
    
    if x == 'text_week':
        pd_text_targets = features
    else:
        pd_text_targets[x] = features
```

```python
pd_comp_targets
```

```python
pd_text_targets
```

```python
# Integrate x with each targets
for x in A_comp:    
    pd_comp[x] = pd_comp_targets[[x]]
    
```

```python
for x in A_text:
    pd_text[x] = pd_text_targets[[x]]
```

```python
pd_comp
```

```python
pd_text
```

```python
# Create inidivudal dataframe of each target and drop NaN
for x in A_comp:
    tem = B_comp.copy()
    filter_list = tem.append(x)
    
    dict_all[x] = pd_comp.filter(tem, axis=1).dropna(axis=0)
```

```python
for x in A_text:
    tem = B_text.copy()
    filter_list = tem.append(x)
    
    dict_all[x] = pd_text.filter(tem, axis=1).dropna(axis=0)
```
